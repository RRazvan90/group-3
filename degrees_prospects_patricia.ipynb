```python
import pandas as pd
import os
import matplotlib.pyplot as plt
import scipy.stats as st
import numpy as np
```

```python
print(os.getcwd())
```

```python
df_degrees= pd.read_csv(r'../group-3/Resources/degrees-that-pay-back.csv')
```

```python
#cleaing  the degrees data frame.
df_degrees ['Starting Median Salary'] = df_degrees ['Starting Median Salary'].str.replace(',', '').str.replace('.', '').str.replace('$', '')
df_degrees ['Starting Median Salary'] = df_degrees ['Starting Median Salary'].astype(float)

df_degrees ['Mid-Career Median Salary'] = df_degrees ['Mid-Career Median Salary'].str.replace(',', '').str.replace('$', '').str.replace('$', '')
df_degrees ['Mid-Career Median Salary'] = df_degrees ['Mid-Career Median Salary'].astype(float)

df_degrees ['Mid-Career 10th Percentile Salary'] = df_degrees ['Mid-Career 10th Percentile Salary'].str.replace(',', '').str.replace('$', '').str.replace('$', '')
df_degrees ['Mid-Career 10th Percentile Salary'] = df_degrees ['Mid-Career 10th Percentile Salary'].astype(float)

df_degrees ['Mid-Career 25th Percentile Salary'] = df_degrees ['Mid-Career 25th Percentile Salary'].str.replace(',', '').str.replace('$', '').str.replace('$', '')
df_degrees ['Mid-Career 25th Percentile Salary'] = df_degrees ['Mid-Career 25th Percentile Salary'].astype(float)

df_degrees ['Mid-Career 75th Percentile Salary'] = df_degrees ['Mid-Career 75th Percentile Salary'].str.replace(',', '').str.replace('$', '').str.replace('$', '')
df_degrees ['Mid-Career 75th Percentile Salary'] = df_degrees ['Mid-Career 75th Percentile Salary'].astype(float)

df_degrees['Mid-Career 90th Percentile Salary'] = df_degrees['Mid-Career 90th Percentile Salary'].str.replace(',', '').str.replace('$', '').str.replace('$', '')
df_degrees['Mid-Career 90th Percentile Salary'] = df_degrees['Mid-Career 90th Percentile Salary'].astype(float)
```

```python
mean_salary = pay_degrees_2['Mid-Career Median Salary'].mean()
mean_salary
```

```python
mean_national_salary = 54132
```

```python
plt.figure(figsize=(20, 10))
threshold = mean_salary
plt.axhline(y=threshold,linewidth=1, color='red', label = "Average Degree Holder Salary", linestyle='dashed')
plt.axhline(y=mean_national_salary,linewidth=1, color='green', label = "Average National Salary", linestyle='dashed')
plt.bar(pay_degrees['Undergraduate Major'],pay_degrees['Mid-Career Median Salary'])
#ax.axhline(mean_salary, c='r')
plt.xlabel("Degree", fontsize = 15)
plt.ylabel("Mid-Career Median Salary ($)", fontsize = 15)
plt.title('Mid-Career Salary by Degree', fontsize = 15)
leg = plt.legend(loc='upper center')
plt.xticks(rotation=90, fontsize = 15)
plt.show()
```

```python
plt.figure(figsize=(20, 10))
pay_degrees_3 = pay_degrees.copy(deep=True)
pay_degrees_3 = pay_degrees_3.loc[pay_degrees_3['Mid-Career Median Salary']>mean_salary]
threshold = mean_salary
plt.axhline(y=threshold,linewidth=1, color='red', label = "Average Degree Holder Salary", linestyle='dashed')
plt.axhline(y=mean_national_salary,linewidth=1, color='green', label = "Average National Salary", linestyle='dashed')
plt.bar(pay_degrees_3['Undergraduate Major'],pay_degrees_3['Mid-Career Median Salary'], color ='lightcoral')
#ax.axhline(mean_salary, c='r', linestyle='dashed')
plt.xlabel("Degree", fontsize = 15 )
plt.ylabel("Mid-Career Median Salary ($)", fontsize = 15)
plt.title('Mid-Career Salary - Degrees that pay above average', fontsize = 15)
leg = plt.legend(loc='upper center')
plt.xticks(rotation=90, fontsize = 15)
plt.show()
```

```python
plt.figure(figsize=(20, 10))
pay_degrees_4 = pay_degrees.copy(deep=True)
pay_degrees_4 = pay_degrees_4.loc[pay_degrees_4['Mid-Career Median Salary']<mean_salary]
threshold = mean_salary
plt.axhline(y=threshold,linewidth=1, color='red', label = "Average Degree Holder Salary", linestyle='dashed')
plt.axhline(y=mean_national_salary,linewidth=1, color='green', label = "Average National Salary", linestyle='dashed')
plt.bar(pay_degrees_4['Undergraduate Major'],pay_degrees_4['Mid-Career Median Salary'], color ='violet')
#ax.axhline(mean_salary, c='r', linestyle='dashed')
plt.xlabel("Degree", fontsize = 15)
plt.ylabel("Mid-Career Median Salary ($)", fontsize = 15)
plt.title('Mid-Career Salary - Degrees that pay below average', fontsize = 15)
leg = plt.legend(loc='upper center')
plt.xticks(rotation=90, fontsize = 15)
plt.show()
```

```python

pay_degrees_5 = pay_degrees.copy(deep=True)

threshold = mean_salary
#plt.axhline(y=threshold,linewidth=1, color='red', label = "Median Salary")

values = pay_degrees_5['Mid-Career Median Salary']
x = pay_degrees_5['Undergraduate Major']

above_threshold = np.maximum(values - threshold, 0)
below_threshold = np.minimum(values, threshold)

figure, ax = plt.subplots(figsize=(20, 10))
ax.bar(x, below_threshold, color = 'cornflowerblue')
ax.bar(x, above_threshold, color = 'blue', bottom = below_threshold)



ax.axhline(mean_salary, c='r', label = "Average Degree Holder Salary", linestyle='dashed')
plt.axhline(y=mean_national_salary,linewidth=1, color='green', label = "Average National Salary", linestyle='dashed')
plt.xlabel("Degree", fontsize = 15)
plt.ylabel("Mid-Career Median Salary ($)", fontsize = 15)
plt.title('Mid-Career Salary by Degree', fontsize = 15)
leg = plt.legend(loc='upper center')
plt.xticks(rotation=90, fontsize = 15)
plt.savefig(r'../group-3/Mid-Career Salary by Degree')
plt.savefig('Mid-Career Salary by Degree.png', dpi=300, bbox_inches='tight')
plt.show()
```
